```python
# zero shot prompting
# No need to add any extra context
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

response = client.responses.create(
    model="gpt-3.5-turbo",
    input="Q: Roger has 5 tennis balls. He buys 2 more cans of tennis balls. Each can has 3 tennis balls does he have now.\n\n ",
    temperature=0.1,
    top_p=0.1,
    max_output_tokens=256,
)

print(response.output_text)
```
