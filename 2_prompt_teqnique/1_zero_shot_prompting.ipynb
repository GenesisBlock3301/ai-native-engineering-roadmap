```python
# zero shot prompting
# No need to add any extra context
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

response = client.responses.create(
    model="gpt-3.5-turbo",
    input="Classify the text into neutral, negative, or positive \n You are bad boy \n Sentiment: ",
    temperature=0.1,
    top_p=0.1,
    max_output_tokens=256,
)

print(response.output_text)
```

```python
# Few-shot learning
# `Add` some more context and an example.

response = client.responses.create(
    model="gpt-3.5-turbo",
    input="A 'whatpu' is a small, furry animal native to Tanzania. An example of a sentence that uses the word whatpu is:\n We were traveling in Africa and we saw these very cute whatpus.\n To do a 'farduddle' means to jump up and down really fast. An example of a sentence that uses the word farduddle is:\n",
    temperature=0.7,
    top_p=0.1,
    max_output_tokens=256,
)

print(response.output_text)
```
